## Ejercicio 1
### Hacer un programa que escriba un diccionario Python en un archivo csv (crear "temporal.csv"). Después de escribir el archivo CSV (incluyendo la cabecera), leer el archivo CSV y mostrar el contenido (por filas).

```python
import csv
csv_columns = ['id','Columna1', 'Columna2', 'Columna3', 'Columna4']
dict_data = {'id':['1', '2', '3'],
    'Columna1':[23, 35, 44],
    'Columna2':[33, 20, 30],
    'Columna3':[11, 20, 65],
    'Columna4':[59, 16, 42]
     }
csv_file = "temporal.csv" #lo voy a crear
try:
    #vamos a crear el archivo
    with open(csv_file, 'w') as archivocsv:
        writer = csv.DictWriter(archivocsv, fieldnames=csv_columns)
        writer.writeheader()        
        writer.writerow(dict_data)
except IOError:
    print("I/O error")
#
data = csv.DictReader(open(csv_file))
print("archivo CSV como diccionario:\n")
for row in data:
    print(row)
```

## Ejercicio 2
### Escribir un programa que lea cada fila de un archivo csv determinado ('pobl_caYprov.csv') y omita el encabezado del archivo. Imprimir también el número de filas y los nombres de los campos.

```python
import csv
with open('pobl_caYprov.csv') as csvfile:
    data = csv.reader(csvfile, delimiter='\t')
    # la siguiente instrucción es para saltarse la primera fila (el encabezado)
    fields = next(data)
    print('campos:',fields)
    for row in data: # row es una lista con los elementos de la fila correspondiente
        print(', '.join(row)) # concatena los elementos/strings de la lista correspondiente
print(f"\nNº de filas total: {data.line_num}")
print('Los nombres de los campos son:')
print(', '.join(field for field in fields))
```

```python
with open('new_num_veterinarios.json', 'w', encoding="utf-8") as f:
    json.dump(datos, f, ensure_ascii=False, indent=4) # se escribe en el nuevo archivo 'new_num_veterinarios.json'
# El parámetro ensure_ascii=False en la función json.dump es para que los caracteres especiales se codifiquen 
# correctamente. Esto permitirá que las tildes o las ñ, se representen de manera adecuada en el archivo JSON.
```
